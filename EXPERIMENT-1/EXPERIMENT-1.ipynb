##categorical variables into quantitative variables

```python
#METHOD1

# replacing values
df1['Sex'].replace(['male', 'female'],[0, 1], inplace=True)
```

```python
df1
```

```python
#METHOD4


from sklearn.preprocessing import LabelEncoder
# creating instance of labelencoder
labelencoder = LabelEncoder()
# Assigning numerical values and storing in another column
df1['Labelencoding_Embarked'] = labelencoder.fit_transform(df1["Embarked"])
df1
```

```python
df1['Labelencoding_Embarked'].value_counts()
```

```python
df1['Cabin'] = df1['Cabin'].notnull().astype(int)
```

```python
# Assuming your DataFrame is named df

# Perform one-hot encoding on the "Cabin" column
cabin_encoded = pd.get_dummies(df1['Cabin'], prefix='Cabin')

# Concatenate the encoded columns with the original DataFrame
df1 = pd.concat([df1, cabin_encoded], axis=1)

# Drop the original "Cabin" column
df1.drop('Cabin', axis=1, inplace=True)
```
